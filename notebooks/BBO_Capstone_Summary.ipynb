# BBO Capstone Summary Notebook

This notebook loads the stored weekly query history and gives a compact overview of best-so-far performance through Week 11.

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/query_history_weeks_1_11.csv')
df.head()
```

```
best_so_far = []
for fn, g in df.groupby('function'):
    g = g.sort_values('week').copy()
    g['best_so_far'] = g['y'].cummax()
    best_so_far.append(g[['week', 'function', 'y', 'best_so_far']])
bsf = pd.concat(best_so_far, ignore_index=True)
bsf.head()
```

```
fig, axes = plt.subplots(4, 2, figsize=(12, 14), sharex=True)
axes = axes.ravel()
for i, fn in enumerate(sorted(bsf['function'].unique())):
    g = bsf[bsf['function'] == fn]
    axes[i].plot(g['week'], g['y'], marker='o', label='weekly output')
    axes[i].plot(g['week'], g['best_so_far'], linestyle='--', label='best so far')
    axes[i].set_title(f'Function {fn}')
    axes[i].set_xlabel('Week')
    axes[i].set_ylabel('Output')
    axes[i].legend(fontsize=8)
plt.tight_layout()
plt.show()
```

## Interpretation

- Steady upward best-so-far curves indicate successful local refinement.
- Flat or noisy curves indicate harder functions, weak signal, or sparse coverage.
- Because the hidden functions are unknown, these plots are descriptive rather than definitive proofs of convergence.
